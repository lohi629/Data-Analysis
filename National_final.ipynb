```python
p=df['Location'].str.split(",")
city=[]
state=[]
for i in p:
    city.append(i[0])
    state.append(i[1])
df['city']=city
df['state']=state
df[['Name','city','state']]
```

```python
len(df[(df['In-state'].isnull()==False) & (df['Tuition and fees'].isnull()==False)]['Name'])
```

```python
p=df['In-state'].fillna(0)
q=df['Tuition and fees'].fillna(0)
ins=[]
tof=[]
def change(s):
  ins=[]
  for i in s:
    if(i!=0):
      l=i.replace(",","")
      m=l.split("$")
      ins.append(m[1])
    else:
      ins.append(i)
  return ins
```

```python
ins=change(p)
df['ins']=ins
```

```python
tof=change(q)
df['tof']=tof
len(df[(df['tof'].astype(float)-df['ins'].astype(float))<=5000.0]['Name'])
```

```python
df[(df['Rank']>120) & (df['Rank']<170) & (df['tof'].astype(float)>15000)]['Name']
```

```python
temp=df.groupby('city')
li=set(df['city'])
for i in li:
  print(i,"****")
  print(temp.get_group(i)['Name'])
```
