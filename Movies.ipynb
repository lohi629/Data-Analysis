```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/movies_1_.csv")
df=df[df['Release Date']!="TBD"]
df['Release Date']=pd.to_datetime(df['Release Date'])
df
```

```python
from datetime import datetime
y=[]
m=[]
for i in df['Release Date']:
  y.append(i.year)
  m.append(i.month)
df['year']=y
df['month']=m
p=df[df['year']==1995]
p.groupby(['month'])['Title'].count()
```

```python
p=df[(df['year']>=1990) & (df['year']<=1998)]
p.groupby(['year'])['Title'].count()
```

```python
df['Major Genre'].fillna('Adventure')
df.groupby('Major Genre')['Title'].count()
```

```python
df['IMDB Rating'].fillna(0)
df['Rotten Tomatoes Rating'].fillna(0)
b=df[(df['IMDB Rating']>7) & (df['Rotten Tomatoes Rating']>60)]
b=b[['Major Genre','Title']]
b.groupby('Major Genre').agg(list)
```

```python
temp=df[df['Major Genre']=='Action'].sort_values('Release Date')
temp.groupby('Director')['Title'].agg(list)
```

```python
df.sort_values(by=['Major Genre','Director']).groupby('Distributor')['Title'].agg(list)
```

```python
df['Worldwide Gross'].replace('Unknown',np.nan,inplace=True)
df['US Gross'].replace('Unknown',np.nan,inplace=True)
s=df[df['Worldwide Gross'].isnull()==False]
s['Worldwide Gross']=s['Worldwide Gross'].astype(int)
s['US Gross']=s['US Gross'].astype(int)
s['res gross']=s['Worldwide Gross'].sub(s['US Gross'])
print(s[['Title','res gross','IMDB Rating']])
```
